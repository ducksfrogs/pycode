```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = 0
std = 1

X = np.arange(-4,4,0.01)

Y = stats.norm.pdf(X, mean, std)

plt.plot(X,Y)
```
