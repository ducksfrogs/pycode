```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips)
```

```python
g = sns.FacetGrid(row='sex', col='time', hue='smoker', 
                  hue_kws={'marker': ['x', 'o']}, data=tips)
g.map(sns.regplot, 'total_bill', 'tip')
```

```python
sns.violinplot(x='day', y='tip', hue='sex', split=True, data=tips)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.factorplot(x='class', y='survived', col='sex', data=titanic)
```

```python
sns.lmplot(x='fare', y='survived', col='sex', data=titanic, logistic=True)
```
